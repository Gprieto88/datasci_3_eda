```python
import matplotlib.pyplot as plt
import seaborn as sns

# sample blood sugar data
sugar_bs = [130, 95, 108, 111, 99, 100, 97, 123, 107]

# Scatter plot
plt.figure(figsize=(25, 65))
sns.scatterplot(x=range(len(sugar_bs)), y=sugar_bs)
plt.xlabel("patient")
plt.ylabel("Blood Sugar Data")
plt.title("Scatter Plot of Blood Sugar Data")
plt.show()
```
